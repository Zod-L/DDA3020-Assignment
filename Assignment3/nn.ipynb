```python
# Process data
import pandas as pd
import numpy as np

train_data = pd.read_csv("./train.csv").to_numpy()
test_data = pd.read_csv("./test.csv").to_numpy()
X_train = train_data[:, 2:]
Y_train = train_data[:, 1]
X_test = test_data[:, 1:]
Y_test = test_data[:, 0]
```

```python
# Train MLP
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

hidden_layers = [1,2,3]
hidden_nodes = [50, 200, 784]
optimizer = ["sgd", "adam"]

for l in hidden_layers:
    for n in hidden_nodes:
        for opt in optimizer:
            model = MLPClassifier(hidden_layer_sizes=l * [n], solver=opt, verbose=False)
            model.fit(X_train, Y_train)
            plt.figure()
            plt.plot(model.loss_curve_)
            plt.title(f"Loss curve of {l}x{n} with {opt} optimizer")
            plt.xlabel("iteration")
            plt.xlabel("loss")

            pred_train = model.predict(X_train)
            pred_test = model.predict(X_test)
            train_error = (pred_train != Y_train).mean()
            test_error = (pred_test != Y_test).mean()
            print("=================================")
            print(f"number of hidden layers: {l}")
            print(f"number of hidden nodes: {n}")
            print(f"optimizer: {opt}")
            print(f"Training error: {train_error}")
            print(f"Testing error: {test_error}")
            
```
